```python
# Caracterizando con gráfico boxplot
fig = pl.figure(figsize=(4,4))
_ = pl.boxplot(data.Store, vert=False)
pl.xlabel(r'Store')
pl.title('Boxplot Store')
pl.grid()
```

```python
# Caracterizando con diagrama de violín
fig = pl.figure(figsize=(4,4))
_ = pl.violinplot(data.Store, quantiles=[0.25,0.5,0.75])
pl.ylabel(r'Store')
pl.title('Diagrama de Violín Store')
pl.grid()
```

```python
columna = data['Store']
# Calcular el rango intercuartílico (IQR)
Q1 = columna.quantile(0.25)
Q3 = columna.quantile(0.75)
IQR = Q3 - Q1
# Definir los límites inferior y superior para detectar datos atípicos
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
# Encontrar los datos atípicos
datos_atipicos = columna[(columna < limite_inferior) | (columna > limite_superior)]
# Mostrar los datos atípicos
print("Datos atípicos:")
print(datos_atipicos)
```
